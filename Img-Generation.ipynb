```python
import torch
from diffusers import StableDiffusionPipeline, StableDiffusionImg2ImgPipeline
from PIL import Image, ImageDraw, ImageFont

# Check if GPU is available
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load Stable Diffusion Pipeline
pipeline = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4").to(device)
pipeline.enable_attention_slicing()

# Optional: Load img2img pipeline for further customization
img2img_pipeline = StableDiffusionImg2ImgPipeline.from_pretrained("CompVis/stable-diffusion-v1-4").to(device)

# Function to generate images based on prompts
def generate_image(prompt, style="realistic", resolution=(512, 512), custom_text=None, text_position="bottom-right"):
    """
    Generate an image based on the input prompt and optional customizations.

    Args:
        prompt (str): Description of the image to generate.
        style (str): "realistic" or "anime" for style preference.
        resolution (tuple): Resolution of the generated image (width, height).
        custom_text (str): Optional text to add to the image.
        text_position (str): Where to place the text: "top-left", "top-right", "bottom-left", "bottom-right".

    Returns:
        PIL.Image: The generated image.
    """
    # Adjust prompt for style
    if style.lower() == "anime":
        prompt = f"Anime style: {prompt}"
    else:
        prompt = f"Realistic style: {prompt}"

    # Generate image
    generated_image = pipeline(prompt, height=resolution[1], width=resolution[0]).images[0]

    # Add custom text if provided
    if custom_text:
        font = ImageFont.truetype("/content/Verdana_Bold_Italic.ttf", 24)
        draw = ImageDraw.Draw(generated_image)
        text_bbox = draw.textbbox((0, 0), custom_text, font=font)
        text_width = text_bbox[2] - text_bbox[0]  # Calculate width
        text_height = text_bbox[3] - text_bbox[1] # Calculate height
        # Calculate text position
        positions = {
            "top-left": (10, 10),
            "top-right": (resolution[0] - text_width - 10, 10),
            "bottom-left": (10, resolution[1] - text_height - 10),
            "bottom-right": (resolution[0] - text_width - 10, resolution[1] - text_height - 10),
        }
        position = positions.get(text_position.lower(), "bottom-right")
        draw.text(position, custom_text, font=font, fill="white")

    return generated_image

# Example Usage
if __name__ == "__main__":
    prompt = input("Enter your image prompt: ")
    style = input("Enter style (realistic/anime): ").strip().lower()
    custom_text = input("Enter text to include in the image (optional): ")
    text_position = input("Enter text position (top-left, top-right, bottom-left, bottom-right): ")

    # Generate the image
    image = generate_image(prompt, style=style, custom_text=custom_text, text_position=text_position)

    # Save and display the image
    image.save("generated_image.png")
    image.show()
```

```python
import openai
from PIL import Image, ImageDraw, ImageFont
import requests
from io import BytesIO

# Set your OpenAI API key
openai.api_key = "#"

# Function to generate an image based on a prompt
def generate_image(prompt, size="1024x1024"):
    """
    Generate an image based on the given prompt using OpenAI's API.
    """
    try:
        response = openai.Image.create(
            prompt=prompt,
            n=1,
            size=size
        )
        image_url = response['data'][0]['url']
        print(f"Generated Image URL: {image_url}")
        return image_url
    except Exception as e:
        print(f"Error generating image: {e}")
        return None

# Function to add text to an image
def add_text_to_image(image_url, text, position="bottom-right"):
    """
    Add text to the generated image at the specified position.
    """
    try:
        # Download the generated image
        response = requests.get(image_url)
        image = Image.open(BytesIO(response.content))

        # Load a font
        try:
            font = ImageFont.truetype("arial.ttf", 36)  # Use a system font
        except IOError:
            font = ImageFont.load_default()

        draw = ImageDraw.Draw(image)
        text_width, text_height = draw.textsize(text, font=font)

        # Determine position
        if position == "bottom-right":
            x, y = image.width - text_width - 20, image.height - text_height - 20
        elif position == "top-left":
            x, y = 20, 20
        elif position == "center":
            x, y = (image.width - text_width) // 2, (image.height - text_height) // 2
        else:
            x, y = 20, image.height - text_height - 20  # Default bottom-left

        # Add text with a white background
        draw.rectangle(
            [x - 5, y - 5, x + text_width + 5, y + text_height + 5],
            fill="white"
        )
        draw.text((x, y), text, fill="black", font=font)

        # Save and display the modified image
        image.show()
        output_path = "output_image_with_text.png"
        image.save(output_path)
        print(f"Image with text saved at: {output_path}")
        return output_path
    except Exception as e:
        print(f"Error adding text to image: {e}")
        return None

# Main function to generate an image and overlay text
def main():
    # Prompt for image generation
    prompt = input("Enter the prompt for the image: ")

    # Generate the image
    image_url = generate_image(prompt)
    if not image_url:
        print("Image generation failed.")
        return

    # Text overlay details
    text = input("Enter the text to overlay on the image: ")
    position = input("Enter the position for the text (bottom-right, top-left, center): ")

    # Add text to the image
    result_image_path = add_text_to_image(image_url, text, position)
    if result_image_path:
        print(f"Final image saved at: {result_image_path}")
    else:
        print("Failed to add text to the image.")

if __name__ == "__main__":
    main()
```
